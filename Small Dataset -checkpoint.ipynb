```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tk
from matplotlib import cm
import pandas as pd
import time
import copy
np.random.seed(0)
```

```python
# Loading the dataset, and visualizing the data
# The dataset here is the random dataset for predicting the profit in $ (in 10,000's) for the food truck
# based on the population (in 10,000's) in the town.

df = pd.read_csv('dataset.txt',header=None)
df.head(10)
```

```python
data = np.loadtxt('dataset.txt',delimiter=',')
x = data[:,0]
y = data[:,1]

# Training examples 
m = len(y)

# Visualizing the data
plt.plot(x,y,'ro',ms=10,mec='k')
plt.title('Profit vs Population')
plt.xlabel('Population in 10,000s')
plt.ylabel('Profit in 10,000s $')
plt.savefig('dataset.png')
```

# Batch Gradient Descent 

```python
# Function for computing cost attached 
def compute_cost(theta):
    J = (0.5/m)*((X@theta - y).T @ (X@theta - y))
    return float(J)
```

```python
# creating numpy arrays 
x = np.array(x)
y = np.array(y)

# stacking 1 to every feature
X = np.stack([np.ones(m),x],axis=1)
y = y.reshape(m,1)


# Parameters for the model h(x) = theta0 + theta1*x
theta = np.zeros(2).reshape(2,1)
theta[0] = 1;
theta[1] = 1

# Learning rate of the model
alpha = 0.01

# Number of iterations 
itera = 1500
```

```python
# Cost values for every iterations 
J = []

# Start time
start_time = time.time()


# Running the Gradient Descent 
for i in range(itera):
    theta = theta - (alpha/m)*(((X@theta - y).T @ X).T)
    J.append(compute_cost(theta))

# End time
end_time = time.time() - start_time

print('Parameters found from the Batch Gradient Descent\n')
print(f'theta0 is {theta[0]}\ntheta1 is {theta[1]}')


print('\n Time taken by the Batch gradient descent :',end_time)
# Predicted values for each instances 
Y_predicted = X@theta
    
```

```python
# Plotting the cost function for validation 
plt.plot([i for i in range(itera)],J,'b--',linewidth=2.5)
plt.xlabel('Iterations')
plt.ylabel('Cost attached')
plt.title('Squared error cost function plot',loc='right')
# plt.savefig('cost_batch.png')
```

```python
# Plotting the Model 
plt.plot(x,y,'ro',ms=10,mec='k',label='Raw Dataset')
plt.plot(x,Y_predicted,'b--',label='Model')
plt.title('Profit vs Population Model')
plt.xlabel('Population in 10,000s')
plt.ylabel('Profit in 10,000s $')
plt.legend()
# plt.savefig('model_batch.png')
```

```python
# Plotting the surface curve and contour curve for visualizing the Minima
theta0 = np.linspace(-20,20,num=300)
theta1 = np.linspace(-10,10,num=300)
Jvals =  np.zeros((300,300))

for i in range(300):
    for j in range(300):
        theta_ = np.array([theta0[i],theta1[j]])
        Jvals[i][j] = compute_cost(theta_.reshape(2,1))
        
        
fig = plt.figure()
fig.set_size_inches(15.5, 10.5, forward=True)
ax = plt.axes(projection='3d')
ax.contour3D(theta0,theta1,Jvals,1000)
plt.xlabel('theta0 values')
plt.ylabel('theta1 values')
# plt.savefig('surface.png')
```

```python
# plotting the contour 
fig,ax = plt.subplots(constrained_layout=True)
cs = ax.contour(theta0,theta1,Jvals,10,cmap=cm.bone)
plt.plot(theta[0],theta[1],'rx',ms=12)
ax.clabel(cs,inline=1,fontsize=5)
cbar = fig.colorbar(cs)
plt.xlabel('theta0 values')
plt.ylabel('theta1 values')
plt.legend(['Local Minima'])
# plt.savefig('contour.png')
```

# Stochastic Gradient Descent 

```python
# Function for calculating predicted value
h = lambda theta_0,theta_1,x: theta_0 + theta_1*x
```

```python
theta_old = np.array([0.,0.])
theta_new = np.array([1.,1.])

alpha = 0.01 # step size

x = x.reshape(m)
y = y.reshape(m)
x_stacked = np.stack([x,y],axis=1)
np.random.shuffle(x_stacked)

# Start time
start = time.time()
iter_num = 0


s_k = np.array([float("inf"),float("inf")])
cost_list = []

for j in range(10):
    for i in range(m):
        iter_num += 1
        theta_old = theta_new
        
        s_k[0] = (h(theta_old[0],theta_old[1],x_stacked[i][0])-x_stacked[i][1])
        s_k[1] = (h(theta_old[0],theta_old[1],x_stacked[i][0])-x_stacked[i][1])*x_stacked[i][0]
        
        theta_new = theta_old - alpha * s_k
    cost_list.append(compute_cost(theta_new))

            
end = time.time()-start

print("Parameters found from the Stochastic Gradient Descent:")
print("theta_0 =", theta_new[0])
print("theta_1 =", theta_new[1])
print('Time taken for the Stochastic Gradient Descent',end)

Y_predicted = X@theta_new
```

```python
# Plotting the cost function for validation 
plt.plot([i for i in range(len(cost_list))],cost_list,'b--',linewidth=2.5)
plt.xlabel('Iterations')
plt.ylabel('Cost attached')
plt.title('Squared error cost function plot',loc='right')
plt.savefig('cost_sgd.png')
```

```python
# Plotting the Model 
plt.plot(x,y,'ro',ms=10,mec='k',label='Raw Dataset')
plt.plot(x,Y_predicted,'b--',label='Model')
plt.title('Profit vs Population Model')
plt.xlabel('Population in 10,000s')
plt.ylabel('Profit in 10,000s $')
plt.legend()
plt.savefig('model_sgd.png')
```
