```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tk
from matplotlib import cm
import pandas as pd
import time
import copy
np.random.seed(0)
```

```python
f = lambda x: x*2+17+np.random.randn(len(x))*10

x = np.random.random(500000)*100
y = f(x) 

# Training examples 
m = len(y)
print(m)

```

```python
# Visualizing the data
plt.plot(x,y,'ro',ms=10,mec='k')
plt.title('Y-X')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('dataset.png')
```

# Batch Gradient Descent 

```python
# Function for computing cost attached 
def compute_cost(theta):
    J = (0.5/m)*((X@theta - y).T @ (X@theta - y))
    return float(J)
```

```python
# creating numpy arrays 
x = np.array(x)
y = np.array(y)

# stacking 1 to every feature
X = np.stack([np.ones(m),x],axis=1)
y = y.reshape(m,1)


# Parameters for the model h(x) = theta0 + theta1*x
theta = np.zeros(2).reshape(2,1)
theta[0] = 12
theta[1] = 0

# Learning rate of the model
alpha = 0.0001

# Number of iterations 
itera = 2000
```

```python
# Cost values for every iterations 
J = []

# Start time
start_time = time.time()


# Running the Gradient Descent 
for i in range(itera):
    theta = theta - (alpha/m)*(((X@theta - y).T @ X).T)
    J.append(compute_cost(theta))

# End time
end_time = time.time() - start_time

print('Parameters found from the Batch Gradient Descent\n')
print(f'theta0 is {theta[0]}\ntheta1 is {theta[1]}\n')

print('\n Time taken by the Batch gradient descent :',end_time)
# Predicted values for each instances 
Y_predicted = X@theta
    
```

```python
# Plotting the cost function for validation 
plt.plot([i for i in range(itera)],J,'b--',linewidth=2.5)
plt.xlabel('Iterations')
plt.ylabel('Cost attached')
plt.title('Squared error cost function plot',loc='right')
plt.savefig('COST_BATCH.png')
```

```python
# Plotting the Model 
plt.plot(x,y,'ro',ms=10,mec='k',label='Raw Dataset')
plt.plot(x,Y_predicted,'b--',label='Model')
plt.title('Y - X')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.savefig('model_batch.png')
```

# Stochastic Gradient Descent 

```python
# Function for calculating predicted value
h = lambda theta_0,theta_1,x: theta_0 + theta_1*x
cost = lambda theta_0,theta_1, x_i, y_i: 0.5*(h(theta_0,theta_1,x_i)-y_i)**2
```

```python
theta_old = np.array([0.,0.])
theta_new = np.array([12.,0.])

alpha = 0.0001 # step size

x = x.reshape(m)
y = y.reshape(m)
x_stacked = np.stack([x,y],axis=1)
np.random.shuffle(x_stacked)

# Start time
start = time.time()
iter_num = 0


s_k = np.array([float("inf"),float("inf")])
cost_list = []
sum_cost = 0

for j in range(5):
    for i in range(m):
        iter_num += 1
        theta_old = theta_new
        
        s_k[0] = (h(theta_old[0],theta_old[1],x_stacked[i][0])-x_stacked[i][1])
        s_k[1] = (h(theta_old[0],theta_old[1],x_stacked[i][0])-x_stacked[i][1])*x_stacked[i][0]
        theta_new = theta_old - alpha * s_k
        sum_cost += cost(theta_old[0],theta_old[1],x_stacked[i][0],x_stacked[i][1])
        if (i+1) % 10000 == 0:
            cost_list.append(sum_cost/10000.0)
            sum_cost = 0   
   

            
end = time.time()-start

print("Parameters found from Stochastic Gradient Descent:")
print("theta_0 =", theta_new[0])
print("theta_1 =", theta_new[1])
print('Time taken for stochastic gradient descent ',end)

Y_predicted = X@theta_new
```

```python
# Plotting the cost function for validation 
plt.plot([i for i in range(len(cost_list))],cost_list,'b--',linewidth=2.5)
plt.xlabel('Iterations')
plt.ylabel('Cost attached')
plt.title('Squared error cost function plot',loc='right')
plt.savefig('cost_sgd.png')
```

```python
# Plotting the Model 
plt.plot(x,y,'ro',ms=10,mec='k',label='Raw Dataset')
plt.plot(x,Y_predicted,'b--',label='Model')
plt.title('Y-X')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.savefig('model_sgd.png')
```
